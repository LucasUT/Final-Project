```python
import numpy as np                 
import pandas as pd 
import matplotlib.pyplot as plt     
import seaborn as sns 
import geopandas 
```

```python
crime = pd.read_csv('data_crime_1217_clean.csv')
```

```python
crime = crime[crime['Primary Type']!='LIQUOR LAW VIOLATION']
crime = crime[crime['Primary Type']!='ARSON']
crime = crime[crime['Primary Type']!='KIDNAPPING']
crime = crime[crime['Primary Type']!='STALKING']
crime = crime[crime['Primary Type']!='INTIMIDATION']
crime = crime[crime['Primary Type']!='OBSCENITY']
crime = crime[crime['Primary Type']!='HOMICIDE']
crime = crime[crime['Primary Type']!='PUBLIC INDECENCY']
crime = crime[crime['Primary Type']!='CONCEALED CARRY LICENSE VIOLATION']
crime = crime[crime['Primary Type']!='NON-CRIMINAL']
crime = crime[crime['Primary Type']!='OTHER NARCOTIC VIOLATION']
crime = crime[crime['Primary Type']!='NON - CRIMINAL']
crime = crime[crime['Primary Type']!='HUMAN TRAFFICKING']
crime = crime[crime['Primary Type']!='NON-CRIMINAL (SUBJECT SPECIFIED)']
```

# Arrest Rate by Crime Type 

```python
types = crime.groupby(by=['Primary Type']).agg({'Arrest':'sum','ID':'count'})
types['rate'] = types['Arrest']/types['ID']
```

```python
types = types.sort_values('rate', ascending=False)
```

```python
types.reset_index(inplace=True)
```

# Per Capita Income (Com Areas) / Arrest Rate

```python
pci = crime.groupby(by=['PER CAPITA INCOME ']).agg({'Arrest':'sum','ID':'count'}).sort_values('PER CAPITA INCOME ')
pci['rate'] = pci['Arrest']/pci['ID']
pci.head()
```

```python
pci.reset_index(inplace=True)
pci.head()
```

```python
pci.to_csv("FINAL_pci.csv")
```

```python
import statsmodels.formula.api as smf

pci.rename({"PER CAPITA INCOME ": "pci"}, axis = 'columns', inplace=True )
```

```python
pci.columns
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.set_style("darkgrid")
ax = sns.scatterplot(x="rate", y="pci", s=100 ,data=pci)
ax.grid(True,axis='both',linewidth=2)
ax.set_xlabel('Arrest Rate',fontsize=30)
ax.set_ylabel('Per Capita Income',fontsize=30)
```

```python
mod=smf.ols('rate ~ np.power(pci, 2) + pci ',pci).fit()
mod.summary()
```

```python
corr=pci[['pci','rate']].corr()
corr.style.background_gradient()
```
